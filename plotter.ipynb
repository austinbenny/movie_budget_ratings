```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
import matplotlib as mpl
```

```python
OutFile = 'plots'
InFile = 'goodDF.csv'
df=pd.read_csv(InFile)
```

```python
xVals = ['rating','rating','Budget','Budget','rating']
yVals = ['Budget','Cumulative Worldwide Gross','Cumulative Worldwide Gross','Opening Weekend United States','runtimes']
titles = ['Can You Pay For A Good Movie?',\
    'People Pay To Watch Good Movies Right?',\
        'Are Blockbusters Worth It?',\
            'Was It All Hype?',\
                'Do Longer Movies Recieve A Higher Rating?']
saveNames = ['budget-rating','revenue-rating','revenue-budget','opening-budget','rating-runtimes']
colors = ['steelblue','mediumslateblue','salmon','seagreen','darkkhaki']



for i,tit in enumerate(titles):

    fig,ax = plt.subplots(figsize=(16,9))
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    x = df.loc[:,xVals[i]]
    y = df.loc[:,yVals[i]]

    ax.scatter(x,y,alpha=0.4,color=colors[i],label=yVals[i])
    ax.set_xlabel(xVals[i])
    ax.set_ylabel(yVals[i])
    ax.set_title(tit,fontsize='16',pad=25,fontweight='semibold')

    # trendline
    coeffs = np.polyfit(x,y,1)
    f = lambda m,x,b: m*x + b
    ax.plot(x,f(coeffs[0],x,coeffs[1]),linestyle='dashed',color='black',label='Linear Regression')

    ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    if i == 2 or i==3:
        ax.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    ax.legend(frameon=True)
    fig.savefig(OutFile + '/' + str(saveNames[i]) + '.pdf',format='pdf',bbox_inches='tight')



# histogram

fig,ax = plt.subplots(figsize=(16,9))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

bins = np.arange(int(len(pd.unique(df.loc[:,'aspect ratio']))))-0.5


ax.hist(df.loc[:,'aspect ratio'],bins=bins,width=0.8,facecolor='lightseagreen', alpha=0.75)
ax.set_xlabel('Aspect Ratio')
ax.set_ylabel('Occurence')
ax.set_title('Is There Something About The Size Of The Image?',fontsize='16',pad=25,fontweight='semibold')
fig.savefig(OutFile + '/' + str('AR') + '.pdf',format='pdf',bbox_inches='tight')
```
